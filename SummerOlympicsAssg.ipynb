```python
import pandas as pd
```

```python
df = pd.read_csv("summer.csv")
```

```python
df
```

1. In how many cities Summer Olympics is held so far?¶

```python
len(df['City'].unique())
```

2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
Gold_df=df[df["Medal"]=="Gold"]
Gold_df.groupby("Sport").size().sort_values(ascending=False).head()
```

3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby("Sport").size().sort_values(ascending=False).head()
```

Which player has won most number of medals? (Top 5)

```python
df.groupby("Athlete").size().sort_values(ascending=False).head()
```

5. Which player has won most number Gold Medals of medals? (Top 5)

```python
gold_df=df[df["Medal"]=="Gold"]
gold_df.groupby("Athlete").size().sort_values(ascending=False).head()
```

6. In which year India won first Gold Medal in Summer Olympics?

```python
ind_df=df[df["Medal"]=="Gold"]
ind1_df=ind_df[ind_df["Country"]=="IND"]

min(ind1_df["Year"])
```

7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby("Event").size().sort_values(ascending=False).head()
```

8. Which sport is having most female Gold Medalists? (Top 5)

```python
fem_df=df[df["Medal"]=="Gold"]
fem1_df=fem_df[fem_df["Gender"]=="Women"]
fem1_df.groupby("Sport").size().sort_values(ascending=False).head()
```
